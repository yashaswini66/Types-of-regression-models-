```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

# Generate synthetic dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Plotting setup
def plot_model(name, model, X_fit=X, color='blue'):
    y_pred = model.predict(X_fit)
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, color='gray', label='Data')
    plt.plot(X_fit, y_pred, color=color, label=f'{name}')
    plt.title(f'{name} Prediction')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    plt.show()

# 1. Linear Regression
lin = LinearRegression()
lin.fit(X_train, y_train)
plot_model("Linear Regression", lin, X)

# 2. Polynomial Regression (degree=3)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(poly.fit_transform(X_train), y_train)
plot_model("Polynomial Regression (degree=3)", poly_model, poly.fit_transform(X), color='green')

# 3. Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
plot_model("Ridge Regression", ridge, X, color='red')

# 4. Lasso Regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
plot_model("Lasso Regression", lasso, X, color='purple')

# 5. ElasticNet Regression
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train, y_train)
plot_model("ElasticNet Regression", elastic, X, color='orange')

# 6. Support Vector Regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(scaler.transform(X_train), y_train)
plot_model("SVR (RBF Kernel)", svr, X_scaled, color='brown')

# 7. Decision Tree Regression
tree = DecisionTreeRegressor(max_depth=5)
tree.fit(X_train, y_train)
plot_model("Decision Tree Regression", tree, X, color='teal')
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor

# Create dataset
np.random.seed(42)
X = np.sort(5 * np.random.rand(100, 1), axis=0)
y = np.sin(X).ravel() + np.random.normal(0, 0.1, X.shape[0])
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Save function
def save_plot(name, model, X_fit=X, color='blue'):
    y_pred = model.predict(X_fit)
    plt.figure(figsize=(6, 4))
    plt.scatter(X, y, color='gray', label='Data')
    plt.plot(X_fit, y_pred, color=color, label=f'{name}')
    plt.title(f'{name} Prediction')
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()
    plt.grid(True)
    filename = name.lower().replace(" ", "_").replace("(", "").replace(")", "") + ".png"
    plt.savefig(filename, dpi=300, bbox_inches='tight')
    plt.close()
    print(f"✅ Saved: {filename}")

# 1. Linear Regression
lin = LinearRegression()
lin.fit(X_train, y_train)
save_plot("Linear Regression", lin)

# 2. Polynomial Regression (degree=3)
poly = PolynomialFeatures(degree=3)
X_poly = poly.fit_transform(X)
poly_model = LinearRegression()
poly_model.fit(poly.fit_transform(X_train), y_train)
save_plot("Polynomial Regression", poly_model, X_fit=poly.fit_transform(X), color='green')

# 3. Ridge Regression
ridge = Ridge(alpha=1.0)
ridge.fit(X_train, y_train)
save_plot("Ridge Regression", ridge, color='red')

# 4. Lasso Regression
lasso = Lasso(alpha=0.1)
lasso.fit(X_train, y_train)
save_plot("Lasso Regression", lasso, color='purple')

# 5. ElasticNet Regression
elastic = ElasticNet(alpha=0.1, l1_ratio=0.5)
elastic.fit(X_train, y_train)
save_plot("ElasticNet Regression", elastic, color='orange')

# 6. Support Vector Regression
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
svr = SVR(kernel='rbf', C=100, epsilon=0.1)
svr.fit(scaler.transform(X_train), y_train)
save_plot("SVR Regression", svr, X_fit=X_scaled, color='brown')

# 7. Decision Tree Regression
tree = DecisionTreeRegressor(max_depth=5)
tree.fit(X_train, y_train)
save_plot("Decision Tree Regression", tree, color='teal')
```
